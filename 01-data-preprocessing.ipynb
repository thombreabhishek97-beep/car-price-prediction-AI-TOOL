# Findings
1. Datatypes are wrong - year - int, Price - float/int, kms_driven - int
2. company has many wrong values
3. year - has many non-year values
4. Price - comma in every value, many record contains - 'Ask For Price'
5. kms_driven - comma in every value, ' kms' in every value, null values, Some wrong values
6. fuel_type - null values
7. name - contains company name
8. Many duplicate records

```python
#Lets clean 'year' column
#convert datatype of column
df['year'] = pd.to_numeric(df['year'], errors = 'coerce')
```

```python
#Lets remove null value records - year
df.dropna(subset = ['year'], inplace = True)
```

```python
#Convert to int datatype
df['year'] = df['year'].astype(int)
```

```python
df.info()
```

```python
#Lets clean Price column
for i in df.index:
    price = df.loc[i, 'Price']
    price = price.replace(',', '')
    df.loc[i, 'Price'] = price
```

```python
#Lets clean 'Price' column
#convert datatype of column
df['Price'] = pd.to_numeric(df['Price'], errors = 'coerce')
```

```python
#Lets remove null value records - Price
df.dropna(subset = ['Price'], inplace = True)
```

```python
#Lets clean kms_driven column
for i in df.index:
    kms_driven = df.loc[i, 'kms_driven']
    kms_driven = kms_driven.replace(',', '')
    kms_driven = kms_driven.replace(' kms', '')
    df.loc[i, 'kms_driven'] = kms_driven
```

```python
df['kms_driven'] = pd.to_numeric(df['kms_driven'], errors = 'coerce')
```

```python
#Lets remove null value records - year
df.dropna(subset = ['kms_driven'], inplace = True)
```

```python
#Lets work  on fuel_type column
df['fuel_type'].unique()
```

```python
#Lets remove null value records - fuel_type
df.dropna(subset = ['fuel_type'], inplace = True)
```

```python
df.info()
```

```python
#Lets clean company column
df['company'].unique()
```

```python
#Lets clean 'name' column
len(df['name'].unique())
```

```python
for i in df.index:
    name = df.loc[i, 'name']
    company = df.loc[i, 'company']
    name = name.replace(company, '').strip()
    name = " ".join(name.split()[:2])
    df.loc[i, 'name'] = name
```

```python
#Duplicate records
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

```python
#Save dataframe as csv file
df.to_csv('cleaned_data.csv', index = False)
```
